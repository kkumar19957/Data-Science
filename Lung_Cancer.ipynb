```python
data=data.drop_duplicates()
```

```python
cols=['SMOKING','YELLOW_FINGERS','ANXIETY','PEER_PRESSURE','CHRONIC DISEASE','WHEEZING','ALCOHOL CONSUMING','COUGHING',
      'SHORTNESS OF BREATH','SWALLOWING DIFFICULTY','CHEST PAIN']
data[cols]=data[cols].replace({
    2:1,
    1:0
})
```

```python
data.columns=data.columns.str.lower()
data.head()
```

```python
data['lung_cancer']=data['lung_cancer'].map({'YES':1,'NO':0})
data.gender=data.gender.map({"M":1,"F":0})
```

```python
data.head()
```

```python
X=data.drop('lung_cancer',axis=1)
y=data.lung_cancer
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
len(X_train),len(X_test),len(y_train),len(y_test)
```

```python
clf=RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
def evaluation(yt,yp):
  """
  Give full evaluation of classification
  define yt=y_test,yp=y_preds
  """
  acc=round(accuracy_score(yt,yp)*100,2)
  f1=round(f1_score(yt,yp)*100,2)
  re=round(recall_score(yt,yp)*100,2)
  pr=round(precision_score(yt,yp)*100,2)
  print(f"Accuracy = {acc}%\nF1 Score = {f1}%\nRecall = {re}%\nPrecision = {pr}%")
```

```python
y_preds=clf.predict(X_test)
y_preds[:5]
```

```python
y_test[:5]
```

```python
evaluation(y_test,y_preds)
```

```python
# Function use to plot confusion metrix
def plot_cm(yt,yp,title):
  """
  define yt=y_test
  yp=y_preds
  title=title of plot

  """
  cm=confusion_matrix(yt,yp)
  sns.heatmap(cm,annot=True,fmt='d')
  plt.title(title)
  plt.xlabel('Predicted')
  plt.ylabel('Actual')
```

```python
plot_cm(y_test,y_preds,"HeatMap Using Random Forest Classifier")
```

```python
grid={
      'max_depth':[2,6,8,None],
      'min_samples_split':[2,6,8],
      'min_samples_leaf':[3,6,9],

     }
scoring = {'accuracy', 'precision', 'recall', 'f1', 'roc_auc'}
```

In the context of predicting cancer, recall is considered an important metric because it focuses on the ability of the model to correctly identify all actual cases of cancer, minimizing the risk of false negatives.

```python
#focus on recall
crv=GridSearchCV(clf,grid,scoring=scoring, cv=5,refit='recall',verbose=1)
```

```python
crv.fit(X_train,y_train)
```

```python
crv.best_estimator_
```

```python
crv.best_params_
```

```python
crv.score(X_test,y_test)
```

```python
y_predsc=crv.predict(X_test)
```

```python
evaluation(y_test,y_predsc)
```

```python
plot_cm(y_test,y_predsc,'HeatMap Using Random Forest After Hyperparameter Tuning')
```
