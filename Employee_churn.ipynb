## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import f1_score,precision_score,recall_score,accuracy_score,confusion_matrix,roc_auc_score
def evaluation(yt,yp,mname):
  '''
  Takes yt=ytest,yp=ypreds and calculate all the metrics
  '''
  acc=round(accuracy_score(yt,yp),2)*100
  pre=round(precision_score(yt,yp),2)*100
  re=round(recall_score(yt,yp),2)*100
  f1=round(f1_score(yt,yp),2)*100
  auc = round(roc_auc_score(yt, yp),2)*100

  table = pd.DataFrame()
  table = pd.DataFrame({'model': [mname],
                          'precision': [pre],
                          'recall': [re],
                          'F1': [f1],
                          'accuracy': [acc],
                          'auc': [auc]
                        })
  return table
```

```python
def plot_cm(yt,yp,title=None):
  '''
  Takes yt=ytest,yp=ypreds and calculate confusion metrics and plot it n heatmap
  Provide title to be shown
  '''
  cm=confusion_matrix(yt,yp)
  sns.heatmap(cm,annot=True,fmt='d')
  plt.title(title,fontsize=10)
  plt.xlabel('Predicted')
  plt.ylabel('Actual')
```

```python
y_preds=lr.predict(X_test)
```

```python
lr_result=evaluation(y_test,y_preds,'Logistic Regression')
```

```python
lr_result
```

```python
plot_cm(y_test,y_preds,'Logistic Regression Heatmap')
```
